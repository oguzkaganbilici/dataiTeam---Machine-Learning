```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Prepare the Data

```python
#Clus-1
x1 = np.random.normal(25,5, 100)
# mean = 25, sigma = 5 , count = 100
y1 = np.random.normal(25, 5, 100)

#Clus-2
x2 = np.random.normal(55,5, 100)
# mean = 25, sigma = 5 , count = 100
y2 = np.random.normal(60, 5, 100)

#Clus-3
x3 = np.random.normal(55,5, 100)
# mean = 25, sigma = 5 , count = 100
y3 = np.random.normal(15, 5, 100)
```

```python
x = np.concatenate((x1, x2, x3), axis = 0 )
y = np.concatenate((y1, y2, y3), axis = 0)
```

```python
datas = {"x": x, "y":y}

dt = pd.DataFrame(datas)
dt.head()
```

# Visualize the Data

```python
plt.scatter(x1,y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)
```

# Create the Dendogram

```python
from scipy.cluster.hierarchy import linkage, dendrogram

merg = linkage(dt, method="ward") # ward cluster iindeki yaylmlar minimize eder
dendrogram(merg, leaf_rotation = 90)
plt.xlabel("data points")
plt.ylabel("Euclidean Distance")
plt.show()
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering

hiearchical = AgglomerativeClustering(n_clusters = 3, affinity="euclidean", linkage="ward")
clusters = hiearchical.fit_predict(dt)
```

```python
dt["clusters"] = clusters
```

```python
plt.scatter(dt.x[dt.clusters == 0], dt.y[dt.clusters == 0], color="red")
plt.scatter(dt.x[dt.clusters == 1], dt.y[dt.clusters == 1], color="green")
plt.scatter(dt.x[dt.clusters == 2], dt.y[dt.clusters == 2], color="blue")
```
