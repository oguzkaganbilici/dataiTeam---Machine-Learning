```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv("data.csv")
dt.head()
```

## Clean the data

```python
dt.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
dt.head()
```

```python
dt.diagnosis = [1 if each == "M" else 0 for each in dt.diagnosis]
dt.head()
```

## Determine the varies

```python
y = dt.diagnosis.values
x_data = dt.drop(["diagnosis"], axis=1)
```

## Normalization

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

## Train - Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=31)
```

## Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=1000, random_state=1)
rf.fit(x_train, y_train)
```

### Accuracy

```python
print("Accuracy of the model: {}".format(rf.score(x_test, y_test)))
```

## Confusion matrix

```python
y_pred = rf.predict(x_test)
y_true = y_test
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_true, y_pred)
```

### Visiualize

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(cm, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("predicted y")
plt.ylabel("true y")
plt.show()
```
